```python
# import dependencies 
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python

# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

#scraping for latest title and paragraph
url = 'https://finance.yahoo.com/cryptocurrencies/?offset=0&count=100'
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html,'html.parser')

# find the class for the the title and paragraph
# article = soup.find_all('a', class_= 'Fw(600)')
   
#     # latest news title
#     news_title = article.find('div', class_= 'content_title').text

#     # latest news title's teaser body
#     news_para = article.find('div', class_= 'article_teaser_body').text

# Quit the browser
browser.quit()
```

```python
article = soup.find_all('a', class_= 'Fw(600)')

article
```

```python
crypto_list = []

for each in article:
    title = each['title']
    symbol = each.text
    crypto_dict = {'Name':title,
                   'Ticker':symbol}
    
    crypto_list.append(crypto_dict)
    
```

```python
crypto_list
```

```python
cryptodf = pd.DataFrame(crypto_list)

cryptodf
```

```python
cryptodf.to_csv('yfcryptolist.csv')
```
