```python
#import csv file with the list of S&P 500 companies
splist = pd.read_csv("sp500_constituents.csv")

splist
```

```python
#s&p500 tickers
sp500 = splist.rename(columns={'Symbol':'Ticker'})

sp500
```

```python
tickerStrings = sp500['Ticker']
```

```python
# yfinance code to separate tickers into pretty dataframe
df_list = list()
for ticker in tickerStrings:
    data = yf.download(ticker, group_by="Ticker", period='365d')
    data['Ticker'] = ticker  # add this column because the dataframe doesn't contain a column with the ticker
    
    df_list.append(data)

# combine all dataframes into a single dataframe
tickerDf = pd.concat(df_list)
```

```python
# reset index to prepare for merge
tickerindex = tickerDf.reset_index()

tickerindex
```

```python
# merge the sp500 df to get add the ticker name
mergeticker = tickerindex.merge(sp500, how = 'right', on = 'Ticker')

mergeticker
```

```python
# dropping columns
closeticker = mergeticker.drop(columns =['High','Low','Open','Adj Close','Sector'])

cleanedticker = closeticker.set_index('Date')

cleanedticker
```

```python
# cleanedticker["__id"] = 1
cleanedticker.insert(0, '__id', range(0, len(cleanedticker)))
# cleanedticker.reset_index()
# cleanedticker["__id"] = cleanedticker.index
```

```python
cleanedticker
```

```python
with sqlite3.connect('stocks.sqlite') as con:
    cleanedticker.to_sql('stocks', con=con, dtype={'__id': 'INTEGER PRIMARY KEY AUTOINCREMENT'})
```

```python
# importing btc csv file
btc = pd.read_csv("data_crypto/bitcoin_2012-1-1_2022-6-27.csv")
btc
```

```python
# add ticker name to corresponding column
btc["Ticker"]='BTC'
# add name column to match with the stock df
btc['Name'] = 'Bitcoin'

btc
```

```python
# set index to date w/ datetime format
btc = btc.set_index(pd.to_datetime(btc['Date']).dt.date)

btc
```

```python
# dropping columns to match with stocks df
btc_cleaned = btc.drop(columns=['High','Low','Open','Market Cap','Date'])

btc_cleaned.head(10)
```

```python
# export to sql
btc_cleaned.to_sql('stocks', conn, if_exists='append')
```
